```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Exercise 1**:

```python
A = np.array([[0.3, 0.6, 0.1],
              [0.5, 0.2, 0.3],
              [0.4, 0.1, 0.5]])

v = np.array([1/3] * 3)
```

```python
dists = []
```

```python
for _ in range(25):
    v_prime = v.dot(A)
    dists.append(np.linalg.norm(v_prime - v))
    v = v_prime
```

```python
plt.plot(dists)

plt.show()
```

```python
v
```

`v` became the eigenvector of A, with respect to the eigenvalue $1$.

$Av^* = v^*$

**Exercise 4**:

```python
rot_digits = []

for i in range(10):
    avg = mnist[mnist.label == i].iloc[:,1:].mean().values.reshape(28,28)
    rot_digits.append(np.rot90(avg, k=-1))
```

**Exercise 5**:

```python
def is_symmetric(M, tol=1e-8):
    return np.all(np.abs(M-M.T) < tol)
```

```python
is_symmetric(np.eye(3))
```

```python
is_symmetric(np.random.random((3,3)))
```

**Exercise 6**:

```python
num_vars = 2
num_points = 3000
```

```python
x = np.random.uniform(low=-1.0, high=1.0, size=(num_points, num_vars))
```

```python
xor = np.logical_xor(x[:,0] > 0, x[:,1] > 0)
```

```python
plt.scatter(x[:,0], x[:,1], c=xor)

plt.show()
```

**Exercise 9**:

```python
df = pd.DataFrame(data=np.hstack((x,xor.reshape(-1,1))), columns=['x1', 'x2', 'XOR'])
```
