```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Some function of time')
plt.title('My cool chart')

plt.show()
```
